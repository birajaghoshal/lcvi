# *Preparation of Last.fm data set*

Remember to download and extract the original data into the right directory.

```python
import pandas as pd
import numpy as np
```

```python
d = pd.read_csv("raw/userid-timestamp-artid-artname-traid-traname.tsv", sep="\t", error_bad_lines=False, header=None)
```

```python
d.rename(columns=dict(enumerate("userid \t timestamp \t musicbrainz-artist-id \t artist-name \t musicbrainz-track-id \t track-name".split(" \t "))), inplace=True)
```

```python
users = d.groupby("userid").count().reset_index().sort_values("artist-name", ascending=False)
users
```

```python
artists = d.groupby("artist-name").count().reset_index().sort_values("userid", ascending=False)
artists.iloc[:100]
```

```python
top_artists = artists.iloc[:1003]["artist-name"]
```

```python
top_users = users.iloc[:1000]["userid"]
```

```python
m2ix = dict((m,ix) for ix, m in enumerate(top_artists))
u2ix = dict((u,ix) for ix, u in enumerate(top_users))

data = np.zeros( (len(u2ix), len(m2ix)), dtype=int)#*float("nan")

for u, m in zip(d["userid"], d["artist-name"]):
    if u in u2ix and m in m2ix:
        data[u2ix[u], m2ix[m]] += 1

print("filled cells: %.2f" % (sum(sum(data!=0)) / (len(m2ix)*len(u2ix))))

data_pd = pd.DataFrame(data)
data_pd.to_csv("lastfm_views_dense.csv", sep=",", header=False, index=False)
```

```python
pd.DataFrame(list(enumerate(top_artists))).to_csv("lastfm_ix2artist.csv", header=False, index=False)
```

```python
pd.DataFrame(list(enumerate(top_users))).to_csv("lastfm_ix2user.csv", header=False, index=False)
```
